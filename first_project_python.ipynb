```python
def get_numbers_sorted():
    numbers = []
    while True:
        num = input("Enter a number (type 'done' to finish): ")
        if num == "done":
            break
        numbers.append(float(num))
    print("Original list: {numbers}")
    print("Ascending order: {sorted(numbers)}")
    print("Descending order: {sorted(numbers, reverse=True)}")
get_numbers_sorted()
```

```python
def modify_fruits(fruits=None):

  if fruits is None:
    fruits = ["apple","mango","banana"]

  print("fruits")

  fruits.append("kiwi")
  print("fruits")

  fruits.insert(2,"grape")
  print("fruits")

  if len(fruits) > 0 and isinstance(fruits[0], str):
    fruits[0] = fruits[0].upper()
  print("After uppercasing first element: fruits")
  if len(fruits) > 1:
    fruits.pop(1)
  print("After pop(1): fruits")
  print("Modification complete.")
print("\n--- Calling modify_fruits() without arguments ---")
modify_fruits()
print("\n--- Calling modify_fruits() with an initial list ---")
my_initial_fruit_list = ["orange", "pear"]
modify_fruits(my_initial_fruit_list)
```

```python
def count_temperatures():
    temps = [25, 32, 18, 42, 12, 28, 35, 22, 31, 14]
    above_30 = 0
    below_15 = 0
    for temp in temps:
        if temp > 30:
            above_30 += 1
        elif temp < 15:
            below_15 += 1
    print("Temperatures: {temps}")
    print("Above 30: {above_30}")
    print("Below 15: {below_15}")

    return above_30, below_15

count_temperatures()
```

```python
def remove_all_occurrences(lst, value):
    new_list = lst.copy()
    while value in new_list:
        new_list.remove(value)
test_list = [1, 2, 3, 2, 4, 2, 5]
print("Original: {test_list}")
print("After removing 2: {remove_all_occurrences(test_list, 2)}")
```

```python
def guessing_game():
    # Secret number
    secret = 42
    attempts = 0

    print("Guess the number between 1 and 100!")

    # Play until guessed
    while True:
        guess = int(input("Your guess: "))
        attempts += 1

        if guess < secret:
            print("Too low")
        elif guess > secret:
            print("Too high")
        else:
            print(f"Correct! in {attempts} attempts")
            break

    return attempts

print("\nQuestion 8:")
guessing_game()
```

```python
def shopping_list():
    shop_list = []
    while True:
        print("\n--- Shopping List ---")
        print("1. Add item")
        print("2. Remove item")
        print("3. Show sorted list")
        print("4. Exit")

        choice = input("Choose (1-4): ")

        if choice == "1":
            item = input("Item name: ")
            shop_list.append(item)
            print("Added {item}")

        elif choice == "2":
            item = input("Item to remove: ")
            if item in shop_list:
                shop_list.remove(item)
                print("Removed {item}")
            else:
                print("Item not found")

        elif choice == "3":
            print("Sorted shopping list:")
            for item in sorted(shop_list):
                print("- {item}")

        elif choice == "4":
            print("Goodbye!")
            break

        else:
            print("Wrong choice")
shopping_list()
```

```python
def cities_length():
    cities = []
    for i in range(7):
        city = input(f"City {i+1}: ")
        cities.append(city)
    print("\nCities and letter count:")
    for city in cities:
        length = len(city.replace(" ", ""))
        print(f"{city} - {length} letters")
cities_length()
```
